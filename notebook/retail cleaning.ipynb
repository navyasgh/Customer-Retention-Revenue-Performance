```python
import pandas as pd
```

```python
df= pd.read_csv('online_retail.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df= df.dropna(subset=['Customer ID'])
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df= df[df['Quantity']>0]
df= df[~df['Invoice'].astype(str).str.startswith('C')]
```

```python
df= df[df['Price']>0]
```

```python
df= df.copy()
```

```python
df['InvoiceDate']= pd.to_datetime(df['InvoiceDate'])
```

```python
df.duplicated().sum()
```

```python
df[df.duplicated()]
```

```python
df= df.drop_duplicates()
```

```python
df.info()
```

```python
df['Revenue']= df['Quantity'] * df['Price']
```

```python
df['Year']= df['InvoiceDate'].dt.year
df['Month']= df['InvoiceDate'].dt.month
df['MonthYear']= df['InvoiceDate'].dt.to_period('M')
```

```python
customer_orders= df.groupby('Customer ID')['Invoice'].nunique().reset_index()
customer_orders.columns= ['Customer ID', 'OrderCount']
```

```python
df.info()
```

```python
customer_orders['CustomerType']= customer_orders['OrderCount'].apply(
    lambda x: 'Repeat' if x>1 else 'New'
)
df= df.merge(customer_orders[['Customer ID', 'CustomerType']], on= 'Customer ID')
```

```python
df['MonthYear']= df['MonthYear'].astype(str)
```

```python
df.rename(columns={'Customer ID':'CustomerID'}, inplace= True)
```

```python
df.to_csv("retail_clean.csv", index=False)
```
